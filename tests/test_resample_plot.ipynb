```python
import lma.resample
import lma.plot
```

```python
dataset = 'rat_retina'

all_trees_sorted = lma.resample.read_dataset(f'datasets/{dataset}.txt')
all_trees_sorted
```

```python
doublet_dict, cell_dict = lma.resample.make_dicts(all_trees_sorted, 'doublet')
doublet_dict, cell_dict
```

```python
df_doublet_motifs = lma.resample.resample_trees_doublets(all_trees_sorted, 
                                                         10000, 
                                                         True, 
                                                         doublet_dict,
                                                         cell_dict)
df_doublet_motifs
```

```python
triplet_dict, doublet_dict, cell_dict = lma.resample.make_dicts(all_trees_sorted, 'triplet')
triplet_dict, doublet_dict, cell_dict
```

```python
df_triplet_motifs = lma.resample.resample_trees_triplets(all_trees_sorted, 
                                                         10000, 
                                                         False, 
                                                         triplet_dict,
                                                         doublet_dict,
                                                         cell_dict)
df_triplet_motifs
```

```python
quartet_dict, doublet_dict = lma.resample.make_dicts(all_trees_sorted, 'quartet')
quartet_dict, doublet_dict
```

```python
df_quartet_motifs = lma.resample.resample_trees_quartets(all_trees_sorted, 
                                                         10000, 
                                                         True, 
                                                         quartet_dict,
                                                         doublet_dict)
df_quartet_motifs
```

```python
(df_true_melt_subset, 
 df_melt_subset, 
 df_melt_100resamples_subset, 
 df_null_zscores_i_c_melt_subset,
 df_null_zscores_i_c_melt_100resamples_subset) = lma.plot.dfs_for_plotting(df_triplet_motifs, 
                                                                           10000, 
                                                                           triplet_dict, 
                                                                           cutoff='auto',
                                                                           num_null=1000)
```

```python
cell_color_dict = lma.plot.make_cell_color_dict(all_trees_sorted, cell_fates=None)
cell_color_dict
```

```python
lma.plot.plot_frequency('triplet', 
                        df_true_melt_subset, 
                        df_melt_subset, 
                        df_melt_100resamples_subset,
                        cell_color_dict,
                        cutoff='auto', 
                        legend_bool=True,
                        legend_pos='outside')
```

```python
lma.plot.plot_deviation('triplet', 
                        df_true_melt_subset, 
                        df_null_zscores_i_c_melt_subset, 
                        df_null_zscores_i_c_melt_100resamples_subset, 
                        cell_color_dict,
                        cutoff='auto', 
                        legend_bool=True,
                        legend_pos='outside')
```
